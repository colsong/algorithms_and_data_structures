```python
# Your answer here
from collections import deque

class TreeNode(object):
    def __init__(self, val = 0, left = None, right = None):
        self.val = val
        self.left = left
        self.right = right

def is_duplicate(root: TreeNode) -> int:
    if not root:
        return -1

    seen_values = set() 
    queue = deque([root]) 

    while queue:
        node = queue.popleft() 

        if node.val in seen_values:
            return node.val
        
        seen_values.add(node.val)
        queue.append(node.left) if node.left else None
        queue.append(node.right) if node.right else None

    return -1
```

```python
# Your answer here
#The function performs a breadth-first search (BFS) using a queue to traverse the tree level by level. 
# Each node is processed once, and each value is checked against a set, which takes O(1) time on average. 
# Since we visit every node once,the overall time complexity is O(n).
#The function uses two main data structures:

#A set to store values of nodes encountered, which in the worst case can contain all n unique values.
#A queue to manage BFS traversal. In the worst case, the queue holds approximately n/2 nodes at the last level, making its maximum size O(n).
#Thus, in the worst case, the function requires O(n) space to store node values and BFS traversal data.
```
